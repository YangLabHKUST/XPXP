```python
def PlotCorrelatin(tme_corr,tme_corr_pvalue,label,ismask=False):
    tme_corr_anno = tme_corr.round(3).astype('str').values
    star_index = tme_corr_pvalue<=(0.05)
    tme_corr_anno[star_index] += ' *'
    star_star_index = tme_corr_pvalue<=(0.05/(tme_corr_pvalue.shape[0]))
    tme_corr_anno[star_star_index] += '*'
    sns.set_context('paper',font_scale=1.8) 
    plt.figure(figsize=(18,18),dpi=200)
    if ismask:
        mask = np.tril(np.ones_like(tme_corr, dtype=np.bool))
        ax = sns.heatmap(tme_corr.T, mask=mask.T, square = True, linewidths = 1,
                              cmap='coolwarm', center=0,alpha=0.9,
            cbar_kws={"shrink": 0.7,'label':label,"orientation": "vertical"},
                         annot=tme_corr_anno.T, fmt='',annot_kws={"size": 16})
    else:
        ax = sns.heatmap(tme_corr.T,square = True, linewidths = 1,
                              cmap='coolwarm', center=0,alpha=0.9,
            cbar_kws={"shrink": 0.7,'label':label,"orientation": "vertical"},
                         annot=tme_corr_anno.T, fmt='',annot_kws={"size": 16})
    ax.tick_params(axis='y',rotation=0)
    ax.tick_params(axis='x',rotation=90)
    ax.tick_params(axis='both', which='both', length=0)
    plt.show()
    
def cov_to_corr(df):
    columns = df.columns
    indexs = df.index
    values = df.values
    corr = np.zeros((df.shape[0],df.shape[1]))
    for i in range(df.shape[0]):
        for j in range(i+1,df.shape[0]):
            corr[i,j] = values[i,j]/(values[i,i]*values[j,j])**.5
    corr = corr.T+corr
    return pd.DataFrame(corr,columns=columns,index=indexs)


def get_quantile_preva_and_ratio(df,trait): 
    prevas = []
    ratios = []
    for i in np.array_split(df,nqs):
        prevas.append(i[trait].sum()/i.shape[0])
        ratios.append(i[trait].sum()/(i.shape[0]-i[trait].sum()))
    return np.array(prevas),np.array(ratios)

    
def PlotROC2(res,title,aucs):
    sns.set_context('paper',font_scale=1.2) 
    fig, ax = plt.subplots(1,1,figsize=(8,8), dpi=190)
    colors = ['C'+str(_) for _ in [8,9,3,5,6,4,7]]
    labels = ['LDpred-inf(EAS)','smtpred(EAS)','xpxp(EAS)','LDpred-inf(EUR)','xpxp','smtpred(EAS+EUR)','xpxp(EAS+EUR)']
    for idx,i in enumerate(res):
        ax.plot(i[0],i[1],color=colors[idx],label=labels[idx]+' AUC:'+str(round(aucs[idx],3)))
    plt.plot([0, 1], [0, 1], color='grey', lw=.8, linestyle='--')
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(title)
    plt.legend(loc="upper left")
    plt.show()
    
def get_res_xpxp3(r1,pgera,le):
    r1 = r1.merge(le,left_on='IID',right_on='IID',how='inner')
    r1['SCORESUM'] = r1.apply(lambda x:x['SCORESUM_x']+x['SCORESUM_y'],axis=1)
    r1_roc = roc_curve(r1[pgera],r1['SCORESUM'])[:2]
    r1_auc = auc(r1_roc[0],r1_roc[1])
    r1_corr = pearsonr(r1['SCORESUM'],r1[pgera])[0]
    r1 = r1.sort_values('SCORESUM')
    r1_preval,r1_ratio = get_quantile_preva_and_ratio(r1,pgera)
    return r1_roc,r1_auc,r1_corr,r1_preval,r1_ratio
    
def preval_to_or(preval):
    oddsrs = []
    for i in range(1,len(preval)):
        oddsr = (preval[i:].mean()/(1-preval[i:].mean()))/(preval[:i].mean()/(1-preval[:i].mean()))
        oddsrs.append(oddsr)
    return oddsrs

def get_odds_ratio(r6,pgera):
    odds_ratio = np.zeros((19,2))
    for i,q in enumerate(np.linspace(0.05,0.95,19)):
        x_quantil = [0 for _ in range(int(r6.shape[0]*q))]+[1 for _ in range(int(r6.shape[0]*q),r6.shape[0])]
        log_reg = sm.Logit(r6[pgera].values, sm.tools.tools.add_constant(x_quantil)).fit(disp=False)  
        beta = log_reg.params[1]
        beta_var = log_reg.cov_params()[1,1]
        odr = np.exp(beta)
        odr_se = (odr**2*beta_var)**.5
        odds_ratio[i] = [odr,odr_se]
    return odds_ratio

def get_res_xpxp4(r1,pgera,le):
    r1 = r1.merge(le,left_on='IID',right_on='IID',how='inner')
    r1['SCORESUM'] = r1.apply(lambda x:x['SCORESUM_x']+x['SCORESUM_y'],axis=1)
    r1_roc = roc_curve(r1[pgera],r1['SCORESUM'])[:2]
    r1_auc = auc(r1_roc[0],r1_roc[1])
    r1_corr = pearsonr(r1['SCORESUM'],r1[pgera])[0]
    r1 = r1.sort_values('SCORESUM')
    r1_preval,r1_ratio = get_quantile_preva_and_ratio(r1,pgera)
    r1_odds_ratio = get_odds_ratio(r1,pgera)
    return r1_roc,r1_auc,r1_corr,r1_preval,r1_odds_ratio

def jackknife_auc_sd(a,b):
    data = []
    for i in range(a.shape[0]):
        roc = roc_curve(np.hstack((a[:i],a[i+1:])),np.hstack((b[:i],b[i+1:])))[:2]
        data.append(auc(roc[0],roc[1]))
    data = np.array(data)
    n = a.shape[0]
    return (((data-data.mean())**2).sum()*(n-1)/n)**.5

def get_res_xpxp5(r1,pgera):
    r1_roc = roc_curve(r1[pgera],r1['SCORESUM'])[:2]
    r1_auc = [auc(r1_roc[0],r1_roc[1]),jackknife_auc_sd(r1[pgera],r1['SCORESUM'])]
    r1_corr = pearsonr(r1['SCORESUM'],r1[pgera])[0]
    r1 = r1.sort_values('SCORESUM')
    r1_preval,r1_ratio = get_quantile_preva_and_ratio(r1,pgera)
    r1_odds_ratio = get_odds_ratio(r1,pgera)
    return r1_roc,r1_auc,r1_corr,r1_preval,r1_odds_ratio

def get_res_smtpred2(r1,pgera,peas):
    r1_roc = roc_curve(r1[pgera],r1[peas])[:2]
    r1_auc = [auc(r1_roc[0],r1_roc[1]),jackknife_auc_sd(r1[pgera],r1[peas])]
    r1_corr = pearsonr(r1[peas],r1[pgera])[0]
    r1 = r1.sort_values(peas)
    r1_preval,r1_ratio = get_quantile_preva_and_ratio(r1,pgera)
    r1_odds_ratio = get_odds_ratio(r1,pgera)
    return r1_roc,r1_auc,r1_corr,r1_preval,r1_odds_ratio
    
    
def ExtractRes5(p_key,peas,peur,pgera,psmt,xpxp_res_path,smtpred_res_path,smtpred_res_path_all,pheno):
    r1 = pd.read_csv(xpxp_res_path+peas+'-EAS.profile',delim_whitespace=True)
    r1 = r1.merge(pheno[['SUBJECT_ID',pgera]],left_on='IID',right_on='SUBJECT_ID',how='inner')
    r1_roc,r1_auc,r1_corr,r1_preval,r1_ratio = get_res_xpxp5(r1,pgera)

    r9 = pd.read_csv(smtpred_res_path+psmt+'-wMT-SBLUP.score',delim_whitespace=True)
    r9 = r9.merge(pheno[['SUBJECT_ID',pgera]],left_on='IID',right_on='SUBJECT_ID',how='inner')
    r9_roc,r9_auc,r9_corr,r9_preval,r9_ratio = get_res_smtpred2(r9,pgera,peas+'-EAS')
    
    r2 = pd.read_csv(smtpred_res_path_all+psmt+'-wMT-SBLUP.score',delim_whitespace=True)
    r2 = r2.merge(pheno[['SUBJECT_ID',pgera]],left_on='IID',right_on='SUBJECT_ID',how='inner')
    r2_roc,r2_auc,r2_corr,r2_preval,r2_ratio = get_res_smtpred2(r2,pgera,peas+'-EAS')

    r3 = pd.read_csv(xpxp_res_path+peas+'-EAS-xpxpeas.profile',delim_whitespace=True)
    r3 = r3.merge(pheno[['SUBJECT_ID',pgera]],left_on='IID',right_on='SUBJECT_ID',how='inner')
    r3_roc,r3_auc,r3_corr,r3_preval,r3_ratio = get_res_xpxp5(r3,pgera)

    r4 = pd.read_csv(xpxp_res_path+peur+'-EUR.profile',delim_whitespace=True)
    r4 = r4.merge(pheno[['SUBJECT_ID',pgera]],left_on='IID',right_on='SUBJECT_ID',how='inner')
    r4_roc,r4_auc,r4_corr,r4_preval,r4_ratio = get_res_xpxp5(r4,pgera)

    r5 = pd.read_csv(xpxp_res_path+peas+'-EAS-xpxp.profile',delim_whitespace=True)
    r5 = r5.merge(pheno[['SUBJECT_ID',pgera]],left_on='IID',right_on='SUBJECT_ID',how='inner')
    r5_roc,r5_auc,r5_corr,r5_preval,r5_ratio = get_res_xpxp5(r5,pgera)
    
    r6 = pd.read_csv(xpxp_res_path+peas+'-EAS-xpxpall.profile',delim_whitespace=True)
    r6 = r6.merge(pheno[['SUBJECT_ID',pgera]],left_on='IID',right_on='SUBJECT_ID',how='inner')
    r6_roc,r6_auc,r6_corr,r6_preval,r6_ratio = get_res_xpxp5(r6,pgera)
    
    #r7 = pd.read_csv(smtpredall_res_path+psmt+'-wMT-SBLUP.score',delim_whitespace=True)
    #r7 = r7.merge(pheno[['SUBJECT_ID',pgera]],left_on='IID',right_on='SUBJECT_ID',how='inner')
    #r7_roc = roc_curve(r7[pgera],r7[peas+'-EAS'])[:2]
    #r7_auc = auc(r7_roc[0],r7_roc[1])

    #res_roc[p_key] = [r1_roc,r2_roc,r3_roc,r4_roc,r6_roc,r7_roc,r5_roc]
    #res_auc[p_key] = [r1_auc,r2_auc,r3_auc,r4_auc,r6_auc,r7_auc,r5_auc]
    res_roc[p_key] = [r1_roc,r9_roc,r3_roc,r4_roc,r5_roc,r2_roc,r6_roc]
    res_auc[p_key] = [r1_auc,r9_auc,r3_auc,r4_auc,r5_auc,r2_auc,r6_auc]
    res_corr[p_key] = [r1_corr,r9_corr,r3_corr,r4_corr,r5_corr,r2_corr,r6_corr]
    res_preval[p_key] = [r1_preval,r9_preval,r3_preval,r4_preval,r5_preval,r2_preval,r6_preval]
    res_oddsratio[p_key] = [r1_ratio,r9_ratio,r3_ratio,r4_ratio,r5_ratio,r2_ratio,r6_ratio]
```

```python
def PlotROC2(res,aucs):
    fig, ax = plt.subplots(1,1,figsize=(8,8), dpi=200)
    #colors = '#348ABD,#FF66FF,#988ED5,#777777,#8EBA42,#FFB5B8,#FBC15E,#E24A33'.split(',')
    colors = '#348ABD,#8EBA42,#E24A33'.split(',')
    labels = labels_ORs[2:]
    for idx,i in enumerate(res):
        if idx%2==0:
            linestyle = ':'
            lw=0.5
        else:
            linestyle = '-'
            lw=0.8
        ax.plot(i[0],i[1],color=colors[idx//2],label=labels[idx],lw=lw,linestyle=linestyle) 
    ax.plot([0, 1], [0, 1], color='grey', lw=1, linestyle='--')
    plt.xlim(0,1)
    plt.ylim(0,1)
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    leg = plt.legend(loc="upper left",fontsize=9)
    for legobj in leg.legendHandles:
        legobj.set_linewidth(1.5)
        
        
    axins = zoomed_inset_axes(ax, 1.7, bbox_to_anchor=(0.96,0.67), bbox_transform=ax.transAxes) # zoom = 6
    for idx,i in enumerate(res):
        if idx%2==0:
            linestyle = '--'
            lw=0.5
        else:
            linestyle = '-'
            lw=1
        axins.plot(i[0],i[1],color=colors[idx//2],label=labels[idx],lw=lw,linestyle=linestyle)
    axins.plot([0, 1], [0, 1], color='grey', lw=1, linestyle='--')
    #axins.set_xtick()
    # sub region of the original image
    x1, x2, y1, y2 = 0, 0.22, 0, 0.36
    axins.set_xlim(x1, x2)
    axins.set_ylim(y1, y2)
    axins.set_xticks([])
    axins.set_yticks([])
    plt.xticks(visible=False)
    plt.yticks(visible=False)

    # draw a bbox of the region of the inset axes in the parent axes and
    # connecting lines between the bbox and the inset axes area
    mark_inset(ax, axins, loc1=2, loc2=4, fc="none", ec="0.5", lw=.7,ls='-.')

    plt.show()
```
